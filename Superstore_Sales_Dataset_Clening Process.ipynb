```python
# Convert relevant columns to appropriate data types
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
## Standaralize Column Names:
# replace space in column headers with _
df.columns = df.columns.str.replace(' ', '_')
# replace - in column headers with _
df.columns = df.columns.str.replace('-', '_')
# Check Changes
df.info()
```

```python
# check number of NaN values
df.isna().sum()
```

```python
# inspecting row with the NaN value
df[df.isna().any(axis=1)]
```

##finding:
all NaN Values are for the same city (Burlington) and Satete (Vermont).

After manualy searching for postal code for this area we found their values are between (05401: 05408)

So we are going to replace NaN values in Postal_Code with first value (05401)

```python
#replacing nan values with 05401
df['Postal_Code'].fillna(value='05401', inplace=True)
```

```python
# Check Changes
df.info()
```

```python
# show duplicate rows
df.duplicated()
```

```python
# Standardize 'Ship Mode' categories
df['Ship_Mode'] = df['Ship_Mode'].str.strip().str.title()

# Standardize 'Country' categories
df['Country'] = df['Country'].str.strip().str.title()
```
